##### A taxa de variação obitosAcumulado / casosAcumulado representa a proporção de óbitos em relação aos casos acumulados em uma determinada região ou população em um dado momento.

##### Essa proporção pode ser utilizada como uma medida de gravidade da doença ou da efetividade das medidas de controle adotadas. Quanto maior a proporção, maior é a gravidade da doença e maior é a necessidade de medidas de controle mais efetivas.

##### Por exemplo, se a taxa de variação obitosAcumulado / casosAcumulado está aumentando ao longo do tempo, isso pode indicar um agravamento da situação da pandemia e a necessidade de reforçar as medidas de prevenção e controle. Por outro lado, se essa taxa está diminuindo ao longo do tempo, isso pode indicar uma melhora na situação da pandemia e a possibilidade de flexibilizar as medidas de controle de forma gradual e responsável.

```python
# Converter a coluna de data para o formato de data do Pandas
df['data'] = pd.to_datetime(df['data'])

# Ordenar o dataframe pela coluna de data
df = df.sort_values('data')

# Calcular a taxa de variação das mortes novas (porcentagem) em relação ao tempo
df['taxa_variacao_obitos'] = df['obitosAcumulado'].diff() / df['casosAcumulado'].diff()
```

```python
#preenchendo com 0 os valores NaN
df = df.fillna(0)
```

```python
# Plotar um gráfico das mortes novas em relação ao tempo
df.plot(x='data', y='obitosNovos', kind='line', figsize=(10, 5))
```

```python
# Plotar um gráfico da taxa de variação das mortes novas em relação ao tempo
df.plot(x='data', y='taxa_variacao_obitos', kind='line', figsize=(10, 5))
```

```python
# Definir as condições iniciais
t = df['data']
y = df['obitosNovos']
y_rk4 = [y[0]] # Condição inicial para y_euler

# Definir o tamanho do passo de tempo
h = 1

# Definir o tempo final
tmax = df['data'].iloc[-1]

# Aplicar o método de rk4 para calcular a evolução temporal das mortes novas
for i in range(len(t)-1):
    k1 = (df['obitosNovos'][i] * df['taxa_variacao_obitos'][i])
    k2 = (df['obitosNovos'][i] * df['taxa_variacao_obitos'][i] ) + (h/2) * k1
    k3 = (df['obitosNovos'][i] * df['taxa_variacao_obitos'][i]) + (h/2) * k2
    k4 = (df['obitosNovos'][i] * df['taxa_variacao_obitos'][i]) + h * k3
    y_rk4.append(y_rk4[i] + (h/6) * (k1 + (2*k2) + (2*k3) + k4))

# Plotar o gráfico das mortes novas e da solução numérica
plt.plot(t, y, label='obitosNovos')
plt.plot(t, y_rk4, label='RK4')
plt.legend()
plt.show()
```

```python
# Definir as condições iniciais
t = df['data']
y = df['obitosNovos']
y_euler = [y[0]] # Condição inicial para y_euler

# Definir o tamanho do passo de tempo(um dia)
h = 1

# Definir o tempo final
tmax = df['data'].iloc[-1]

# Aplicar o método de Euler_Melhorado para calcular a evolução temporal das mortes novas
for i in range(len(t)-1):
        y_euler.append(y_euler[i] + (h/2) * (df['obitosNovos'][i] * df['taxa_variacao_obitos'][i]))

# Plotar o gráfico das mortes novas e da solução numérica
plt.plot(t, y, label='obitosNovos')
plt.plot(t, y_euler, label='Euler_Melhorado')
plt.legend()
plt.show()
```

```python
# Definir as condições iniciais
t = df['data']
y = df['obitosNovos']
y_euler = [y[0]] # Condição inicial para y_euler

# Definir o tamanho do passo de tempo(um dia)
h = 1

# Definir o tempo final
tmax = df['data'].iloc[-1]

# Aplicar o método de Euler para calcular a evolução temporal das mortes novas
for i in range(len(t)-1):
    if t[i] < tmax: # Condição para não extrapolarmos o tempo final
        y_euler.append(y_euler[i] + h * (df['obitosNovos'][i] * df['taxa_variacao_obitos'][i]))

# Plotar o gráfico das mortes novas e da solução numérica
plt.plot(t, y, label='obitosNovos')
plt.plot(t, y_euler, label='Euler')
plt.legend()
plt.show()
```
